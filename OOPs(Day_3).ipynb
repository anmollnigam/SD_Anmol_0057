```python
users = {
    "anmol": {
        "pin": "1234",
        "balance": 8500,
        "income": 15000,
        "expense": 6500,
        "transactions": [
            ("Salary", "+15000"),
            ("Food", "-1200"),
            ("Shopping", "-3000"),
            ("Recharge", "-500"),
            ("Petrol", "-1800")
        ]
    },

    "rahul": {
        "pin": "5678",
        "balance": 12000,
        "income": 20000,
        "expense": 8000,
        "transactions": [
            ("Salary", "+20000"),
            ("Electricity Bill", "-2500"),
            ("Movie", "-600"),
            ("Groceries", "-2200"),
            ("Fuel", "-2700")
        ]
    }
}

username = input("Enter Username: ")
pin = input("Enter PIN: ")

if username in users and users[username]["pin"] == pin:
    user = users[username]

    print("\n===== Expense Tracker =====")
    print("Username :", username)
    print("Balance  :", user["balance"])
    print("Income   :", user["income"])
    print("Expense  :", user["expense"])

    print("\nTransactions")
    print("--------------------------")
    for t in user["transactions"]:
        print(t[0], t[1])

else:
    print("Invalid Username or PIN")
```
