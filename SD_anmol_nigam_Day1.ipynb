```python
menu = {
    "Cold Tea": 90,
    "Chowmein": 70,
    "Burger": 100
}

print("Welcome to our Restaurant")
print(menu)

order = input("Enter your item: ")

order_amount = 0

if order in menu:
    order_amount = menu[order]
    print("Your total bill =", order_amount)
else:
    print("Item not available")
```
